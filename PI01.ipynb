```python
#Guardamos archivos en variables de tipo dataframe.
df_amazon = pd.read_csv('Datasets/amazon_prime_titles.csv')
df_amazon.head()
```

```python
#Creamos columna con el nombre de cada plataforma para incorporar al Dataframe unificado.
df_amazon['plataform'] = 'Amazon'
df_disney['plataform'] = 'Disney'
df_hulu['plataform'] = 'Hulu'
df_netflix['plataform'] = 'Netflix'
```

```python
#Unimos Dataframes y probamos si hay duplicados eliminarlos
df_total = pd.concat([df_amazon, df_disney, df_hulu, df_netflix],ignore_index=True).drop_duplicates()
```

```python
df_total
```

```python
#Eliminamos columnas que no vamos a utilizar para las consultas que necesitamos.
df_total = df_total.drop(columns=['description', 'country', 'date_added', 'rating', 'show_id'])
```

```python
#Contamos los nulos por filas
print(df_total.isna().sum(axis=1))
```

```python
df_total.rename(columns={'type':'Tipo_film',  # Cambiamos los nombres de las columnas para leer mas claramente la informacion.
                    'title':'Titulo', 
                    'director':'Director',
                    'cast':'Reparto',
                    'release_year':'Ano_realizacion',
                    'duration':'Duracion',
                    'listed_in':'Genero',
                    'plataform':'Plataforma'}, inplace=True)
```

```python
df_total['Indice'] = df_total.index #Asignamos ID unicos con los indices autoincrementables
```

```python
df_total = df_total[['Indice','Titulo','Tipo_film','Ano_realizacion','Reparto','Duracion','Genero','Plataforma']] #Ordenamos las columnas
```

```python
df_dur=df_total["Duracion"].str.split(expand=True) #Creamos un dataframe con los datos de la columna "duracion" separando el nÃºmero de la unidad
df_dur.rename(columns={0:"Duracion",1:"tipo (min-Seasons)"},inplace=True) #Se renombran las columnas.
df_total.drop(columns='Duracion', inplace=True) #Se elimina la columna 'duracion' del dataframe original total.
df_total['Duracion'] = df_dur["Duracion"]   
df_total['tipo (min-Seasons)'] = df_dur['tipo (min-Seasons)']
```

```python
#Se cambian los valores "sin dato" de las unidades de tiempo (minutos o seasons) en la columna correspondiente
df_total['tipo (min-Seasons)'][df_total['tipo (min-Seasons)'].loc[(df_total['Tipo_film']=='TV Show')].index]='season'
df_total['tipo (min-Seasons)'][df_total['tipo (min-Seasons)'].loc[(df_total['Tipo_film']=='Movie')].index] = 'min'
```

```python
#Reemplazamos los nulos por 'Sin Dato' para optimizar la busqueda a excepcion de la columna duracion que reemplazamos por 0.
df_total['Duracion'].fillna(0, inplace=True)
df_total.fillna('Sin Dato', inplace=True)
df_total['Duracion']=df_total['Duracion'].astype('int64') #convertimos los datos de la columna Duracion a tipo entero.
df_total.head(10)
```

```python
# Calculamos el promedio de duracion por plataforma para completar los nulos con valores que no modifiquen las duraciones maximas de cada una filtrando por Plataforma y tipo de film

prom_pelicula = {}
prom_serie = {}
for plataforma in df_total['Plataforma'].unique():
    prom_pelicula[plataforma] = round(df_total[(df_total['Duracion'] !=0 ) & (df_total['Tipo_film']=='Movie') & (df_total['Plataforma'] == plataforma)]['Duracion'].mean())
    prom_serie[plataforma] = round(df_total[(df_total['Duracion'] !=0 ) & (df_total['Tipo_film']=='TV Show') & (df_total['Plataforma'] == plataforma)]['Duracion'].mean())
    df_total['Duracion'][df_total['Duracion'].loc[(df_total['Duracion'] ==0 ) & (df_total['Tipo_film']=='Movie') & (df_total['Plataforma'] == plataforma)].index] = prom_pelicula[plataforma]
    df_total['Duracion'][df_total['Duracion'].loc[(df_total['Duracion'] ==0 ) & (df_total['Tipo_film']=='TV Show') & (df_total['Plataforma'] == plataforma)].index] = prom_serie[plataforma]
print(prom_pelicula, prom_serie)
```

```python
#Se convierten los actores en listas dentro del dataframe
df_total['Reparto_1'] = df_total['Reparto'].str.split("," or ":")
df_total.drop(columns=['Reparto'],inplace=True)
df_total.rename(columns={'Reparto_1':'Reparto'},inplace=True)

#creaciÃ³n de dataframe con todos los actores
actores = []
for cast in df_total['Reparto']:
    for actor in cast:
        actores.append(actor)

df_actors = pd.DataFrame(actores)
df_actors
```

```python
#CreaciÃ³n de DataFrame para los actores
dic_actores = {'actor':[],
                'year':[],
                'plataforma':[]} #Se crea un diccionario para luego ser convertido en dataframe
for index, actors in enumerate(df_total['Reparto']):
    for actor in actors:
        dic_actores['actor'].append(actor) #Se agrega el actor al diccionario
        dic_actores['year'].append(df_total['Ano_realizacion'][index]) #Se agrega el aÃ±o correspondiente actor al diccionario
        dic_actores['plataforma'].append(df_total['Plataforma'][index]) #Se agrega la plataforma correspondiente al actor al diccionario
df_actores = pd.DataFrame(dic_actores)
#df_actors=df_actors.drop_duplicates().reset_index()
```

```python
df_total['Genero_1'] = df_total["Genero"].str.split(",")
df_total.drop(columns=['Genero'],inplace=True)
df_total.rename(columns={'Genero_1':'Genero'}, inplace=True)
```

```python
#CreaciÃ³n de DataFrame para los generos
dic_genero = {'genero':[],
                'plataforma':[]} #Se crea un diccionario para luego ser convertido en dataframe
for index, generos in enumerate(df_total['Genero']):
    for genero in generos:
        dic_genero['genero'].append(genero) #Se agrega el genero al diccionario
        dic_genero['plataforma'].append(df_total['Plataforma'][index])
df_genero = pd.DataFrame(dic_genero)
```

```python
def get_count_plataform(plataforma:str):   
    return{f"Cantidad de pelÃ­culas en {plataforma} es":df_total[(df_total['Plataforma']==plataforma)&(df_total['Tipo_film']=='movie')]['Tipo_film'].count(),
    f"Cantidad de series en {plataforma} es":df_total[(df_total['Plataforma']==plataforma)&(df_total['Tipo_film']=='tv show')]['Tipo_film'].count()}
```

```python
def get_listedin(genero:str):
    list_genero = []
    for plataforma in df_total['Plataforma'].unique():
        list_genero.append(df_genero[(df_genero['genero']==genero)&(df_genero['plataforma']==plataforma)].shape[0])
    
    maximo = max(list_genero)
    if list_genero.index(maximo)==0:
        return {'amazon': maximo}
    elif list_genero.index(maximo)==1:
        return {'disney': maximo}
    elif list_genero.index(maximo)==2:
        return {'hulu': maximo}
    elif list_genero.index('netflix',maximo)==2:
        return {'netflix': maximo}
```

```python
get_listedin('drama')
```

```python
def get_actor(plataforma:str, aÃ±o:int):
    data = df_actores[(df_actores['year']==aÃ±o)&(df_actores['plataforma']==plataforma)]['actor']
    # return {data.value_counts().max()}
    if data.value_counts().index[0] == 'sin dato':
        return {data.value_counts().index[1]:data.value_counts()[1]}
    elif data.value_counts().index[0] != 'sin dato':
        return {data.value_counts().index[0]:data.value_counts()[0]}
```
